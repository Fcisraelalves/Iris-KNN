```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier
from google.colab import files
import numpy as np
import seaborn as sbs
import matplotlib.pyplot as plt
import pandas as pd
import joblib
```

```python
#carregando o dataset e separando features e labels
iris = load_iris()

X = iris.data
y = iris.target
```

```python
#visualização tabular

data = pd.DataFrame(data=X, columns=iris.feature_names)
data["target names"] = pd.Categorical.from_codes(y, iris.target_names)
data["target"] = y
data.head()
```

```python
#visualização gráfica

sbs.pairplot(data.iloc[:, :5], hue="target names")
plt.show()
```

```python
#dividindo o conjunto em teste e treino

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)
knnModel = KNeighborsClassifier(11)
knnModel.fit(X_train, y_train)

y_pred = knnModel.predict(X_test)

resultados = classification_report(y_test, y_pred)
print(resultados)
```

```python
joblib.dump(knnModel, "iris_knn.pkl")
#files.download("iris_knn.pkl") #use essa linha se executar o notebook no colab.
```
